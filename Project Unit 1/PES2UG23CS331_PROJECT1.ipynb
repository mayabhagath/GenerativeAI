```python
from transformers import pipeline
```

```python
ner_pipeline = pipeline(
    "ner",
    model="dslim/bert-base-NER",
    aggregation_strategy="simple"
)
```

```python
def extract_locations(text):
    entities = ner_pipeline(text)

    locations = []

    for ent in entities:
        if ent["entity_group"] in ["LOC", "GPE"]:
            locations.append(ent["word"])

    return list(set(locations))
```

```python
blog = """
Last summer I traveled across Europe starting in Paris.
From there, I visited Rome, Florence, and Venice in Italy.
Later, I explored Barcelona in Spain and ended my journey in Amsterdam.
"""

places = extract_locations(blog)

print(" Locations mentioned:")
for place in places:
    print("-", place)
```
